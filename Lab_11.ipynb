# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1] * (1 << n) for _ in range(n)]
    parent = [[-1] * (1 << n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1 << n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1 << next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1 << next_city))

                if new_cost < ans:
                    ans = new_cost
                    parent[city][mask] = next_city

        dp[city][mask] = ans
        return ans

    def get_path():
        path = [0]
        mask = 1
        city = 0

        while True:
            next_city = parent[city][mask]
            if next_city == -1:
                break
            path.append(next_city)
            mask |= (1 << next_city)
            city = next_city

        path.append(0)
        return path

    min_cost = visit(0, 1)
    path = get_path()

    return min_cost, path



n = int(input("Enter number of cities: "))
print("Enter the distance matrix:")
34
graph = []
for i in range(n):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    graph.append(row)


cost, path = tsp(graph)

print("\nMinimum travelling cost:", cost)
print("Path taken:", " -> ".join(map(str, path)))
```
